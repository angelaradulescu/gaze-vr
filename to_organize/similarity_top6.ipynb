```python
def spearman_correlation(x, y):
    coef, p = spearmanr(x, y)
    return coef, p
```

```python
def avg_spearman_correlation_between_df(df1, df2):
    
    # check if lenght of both the dataframe are equal
    
    spearman_val_total = 0
    if len(df1) == len(df2):
        for row in df1:
            x = df1.iloc[int(row)][:]
            y = df2.iloc[int(row)][:]
            
            spearman_val = spearman_correlation(x,y)[0]
            spearman_val_total += spearman_val
            
        return spearman_val_total/len(df1)
            
        
        
```

```python
avg_spearman_correlation_between_df(color_csv, embedding_csv)
```

```python
def top_6(row):
    
    return row.argsort()[1:7]
  
def bottom_6(row):
    
    return row.argsort()[::-1][:6]

```

```python

def top(row, n = 10):
    
    return row.argsort()[1:n+1]
  
def bottom(row, n = 10):
    
    return row.argsort()[::-1][:n]
```

```python
em = embedding_csv.apply(top_6, axis = 1)
```

```python
sh = shape_csv.apply(top_6, axis = 1)
```

```python
ch = color_csv.apply(top_6, axis = 1)
```

```python
ch.columns = [1,2,3,4,5,6]
```

```python
def hit_ratio(df1, df2):
    hit_ratio_sum = 0
    
    for i in range(len(df1)):
        
        set_of_similar = set(df1.iloc[[i]].values[0]).intersection(set(df2.iloc[[i]].values[0]))
        hit_ratio_sum += len(set_of_similar)
    
    return hit_ratio_sum/len(df1)
```
